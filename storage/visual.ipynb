```python
def get_rating(_id=None):
    agg_func_custom = {
        'result': ['sum', 'count']
    }
    records = pd.read_excel('records.xlsx', index_col=0)
    stat = records.groupby(['id']).agg(agg_func_custom)
    stat['Winrate'] = round(stat['result']['sum'] / stat['result']['count'], 1) * 100
    stat = stat.sort_values(by=['Winrate'], ascending=False)
    #stat.columns = ['Победы', 'Общее число игр', 'Винрейт']
    rate_dict = stat.to_dict()
    print(rate_dict)
    return stat

get_rating()
```
